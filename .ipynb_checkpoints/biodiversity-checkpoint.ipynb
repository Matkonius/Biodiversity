#### Most seen spieces

```python
# creating search function
def most_seen_animal(df):
    # first benchamrk
    animal = df.iloc[0]
    name_animal_benchmark = animal.scientific_name
    observations_benchmark = animal.observations 
    #comparing with benchmark
    for i in range(len(df)):
        animal = df.iloc[i]
        name_animal = animal.scientific_name
        observations = animal.observations
        if observations > observations_benchmark:
            name_animal_benchmark= name_animal
            observations_benchmark = observations
    
    #finding common_name
    for x in range(len(spec)):
        animal_spec = spec.iloc[x]
        scientific_name = animal_spec.scientific_name
        if scientific_name == name_animal_benchmark:
            name_animal_benchmark = animal_spec.common_names
            category = animal_spec.category
            break
    
    #returning values
    return name_animal_benchmark, observations_benchmark, category
```

##### In park
***

```python
# yellowstone
yellowstone = observ[observ.park_name=="Yellowstone National Park"].reset_index(drop=True)
yellowstone_animal = list(most_seen_animal(yellowstone))

# smoky mountains
smoky_mountains = observ[observ.park_name=='Great Smoky Mountains National Park'].reset_index(drop=True)
smoky_mountains_animal= list(most_seen_animal(smoky_mountains))

# yosemite
yosemite = observ[observ.park_name=="Yosemite National Park"].reset_index(drop=True)
yosemite_animal = list(most_seen_animal(yosemite))

#bryce
bryce = observ[observ.park_name=="Bryce National Park"].reset_index(drop=True)
bryce_animal= list(most_seen_animal(bryce))

#creating_df
    #separate list
animals= [smoky_mountains_animal, yosemite_animal, bryce_animal, yellowstone_animal]
animals_names = [x[0] for x in animals]
animals_observations = [x[1] for x in animals]
animals_category = [x[2] for x in animals]
parks = list(observ.park_name.unique())
        #special list for chart
name_park = list(zip(animals_names, parks))
name_park = ["{} \n {}".format(x[0], x[1]) for x in name_park]

    #creating dictonaries
animals_dict = {}
animals_dict["park"] = parks
animals_dict["name"]= animals_names
animals_dict["observation"]= animals_observations
animals_dict["category"]= animals_category
animals_dict["name and park"]= name_park
    #creating df finally
most_seen_animal_df = pd.DataFrame(data= animals_dict)

#creating chart
sns.set_theme(context= "notebook", style= "whitegrid", palette= "colorblind", font_scale= 1.25)
plt.figure(figsize=(18, 6))
plt.title("Most seen spieces in each park")
sns.barplot(data= most_seen_animal_df, x= "name and park", y="observation")
```

**Comentary:** *These are the most observed species in each park. It is a plant for all parks.*

#### Category --> park

```python
def groupby_category(df):
#finding categories
    df= df.merge(spec, on= "scientific_name", how="inner")
#grouping by it
    group = df.groupby("category").observations.sum().reset_index()
    total_observ = observ.observations.sum()
    #group = group.observations.apply(lambda x: x)
    return group
```

```python
yellowstone_groups= groupby_category(yellowstone)
smoky_mountains_groups= groupby_category(smoky_mountains)
yosemite_groups = groupby_category(yosemite)
bryce_groups = groupby_category(bryce)

yellowstone_groups
```

```python
#creating df + chart
    #lists
category_list = ["Amphibian", "Bird", "Fish", "Mammal", "Nonvascular Plant","Reptile", "Vascular Plant"]
yellowstone_list = list(yellowstone_groups.observations)
smoky_list = list(smoky_mountains_groups.observations)
yosemite_list = list(yosemite_groups.observations)
bryce_list= list(bryce_groups.observations)
park1 = ["Yellowstone" for x in range(7)]
park2 = ["Smoky mountains" for x in range(7)]
park3 = ["Yosemite" for x in range(7)]
park4 = ["Bryce" for x in range(7)]
park = park1 + park2 + park3 + park4
    #dict
groups_dict= {}
groups_dict["category"] = category_list + category_list + category_list + category_list
groups_dict["observations"]= yellowstone_list + smoky_list +yosemite_list +bryce_list
groups_dict["National park"] = park
#groups_dict["park"] 
    #df
groups_df = pd.DataFrame(data=groups_dict)
```

```python
plt.clf()
plt.figure(figsize=(18, 6))
plt.title("Total observations - category + park")
sns.barplot(data= groups_df, x="category", y="observations", hue="National park")
```

**Comentary:** *This graph clearly demonstrate that most observed category is VAscular Plant in Yellowstone park.*

```python
for category in category_list:
    df = groups_df[groups_df.category== category]
    total_observ = df.observations.sum()
    df.observations = df.observations.apply(lambda x: x/total_observ)
    df.observations = df.observations.apply(lambda x: x * 100)
    print(df)
    
    
    plt.clf()
    plt.figure(figsize=(18, 6))
    plt.title(category)
    sns.barplot(data= df, x="category", y="observations", hue="National park")
    plt.show()
```

**Comentary:** *Graphs shows that all categories are mostly observed in Yellowstone National Park (over 40 % of observations for each category).*

# Best park for each spiece

```python
#getting ready df
observ_common_name= observ.merge(spec, on= "scientific_name", how="inner")
observ_common_name =  observ_common_name.drop(columns = ["conservation_status"], axis= 1)
```

```python
#creating list
unique_scientific_name= list(observ_common_name.scientific_name.unique())
best_park = []
scientific_name = []
common_names = []
category = []

for name in unique_scientific_name:
    #creating df
    spieces_df = observ_common_name[observ_common_name.scientific_name == name]
    observations_list= list(spieces_df.observations)

    #finding the best
    index= 5
    benchmark = 0
    for i in range(len(observations_list)):
        observations= observations_list[i]
        if observations > benchmark:
            benchmark = observations
            index = i

    #selecting the best
    best_park.append(spieces_df.iloc[index].park_name)
    scientific_name.append(name)
    
    #adding extra informations
    common_names.append(spieces_df.common_names.iloc[0])
    category.append(spieces_df.category.iloc[0])
```

```python
#creating dictonary
best_see_dict = {}
best_see_dict["scientific_name"]= scientific_name
best_see_dict["common_names"]= common_names
best_see_dict["category"]= category
best_see_dict["most_obesrvations_in"] = best_park

#creating df
most_observations_in = pd.DataFrame(data=best_see_dict)
most_observations_in.most_obesrvations_in.unique()

most_observations_in.most_obesrvations_in.value_counts()
```

**Comentary:** *I have created a table which shows best park to go for all spiecies, but best option for each animal or flower is Yellowstone National Park, therefore isn't necessery to export table or create a chart.
